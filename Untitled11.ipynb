```python
from apify_client import ApifyClient
import pandas as pd

client = ApifyClient("apify_api_QgqeP0nhSM1tGssufanCstEXLqTjbj3HOYN0")

# Ubah data di bawah ini (custom parameter)
input_data = {
    "searchTerms": [
        "#kecuranganpemilu",
        "#pemilucurang",
        "#dirtyvote"
    ],
    "maxItems": 10000,  # jumlah data yang di inginkan
}

print("Mulai scraping...")
run = client.actor("CJdippxWmn9uRfooo").call(run_input=input_data)

data = []
for item in client.dataset(run.default_dataset_id).iterate_items():
    # Ekstrak hashtag dari teks
    text = item.get("text", "")
    hashtags_found = [word for word in text.split() if word.startswith("#")]

    data.append({
        "text": text,
        "date": item.get("createdAt", ""), # Ini awalnya hanya mengambil string mentah dari Twitter/Apify
        "username": item.get("author", {}).get("userName", ""),
        "likes": item.get("likeCount", 0),
        "retweets": item.get("retweetCount", 0),
        "hashtags": ", ".join(hashtags_found),  # ← fix kolom hashtags yang tadinya kosong
    })

df = pd.DataFrame(data)

# =========================================================
# PERBAIKAN FORMAT TANGGAL
# =========================================================
# 1. Konversi teks mentah menjadi objek Datetime Pandas (dengan zona waktu awal UTC/GMT+0)
df['date'] = pd.to_datetime(df['date'], utc=True)

# 2. Opsional: Konversi zona waktu ke Waktu Indonesia Barat (WIB / GMT+7)
df['date'] = df['date'].dt.tz_convert('Asia/Jakarta')

# 3. Merapikan format tampilan menjadi format standar "Tahun-Bulan-Tanggal" tanpa waktu
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
# =========================================================

# Menghitung data sebelum penghapusan duplikat untuk akurasi laporan
jumlah_data_awal = len(df)

# Hapus duplikat
df = df.drop_duplicates(subset=["text"])
print(f"✅ Total terkumpul : {len(df)} tweet")
print(f"📅 Rentang tanggal : {df['date'].min()} s/d {df['date'].max()}")

# Fix bug: Rumus duplikat diubah dari "1000 - len(df)" menjadi hitungan dinamis
print(f"🔁 Duplikat dihapus: {jumlah_data_awal - len(df)} tweet")

df.to_csv("dataset_pemilu.csv", index=False)
print("\nFile berhasil disimpan sebagai 'dataset_pemilu.csv'")
```

```python
# ==========================================
# SCRIPT FILTER DATA REDUNDAN (DUPLIKAT)
# UNTUK GOOGLE COLAB
# ==========================================

# Mengimpor library yang dibutuhkan
import pandas as pd
from google.colab import files
import io

# --- SEL 1: MENGUNGGAH FILE CSV ---
print("Silakan unggah file CSV Anda:")
# Memunculkan tombol upload di Google Colab
uploaded = files.upload()

# Mendapatkan nama file yang baru saja diunggah
# Asumsi pengguna hanya mengunggah satu file
nama_file_awal = list(uploaded.keys())[0]
print(f"\nFile '{nama_file_awal}' berhasil diunggah.\n")

# --- SEL 2: MEMBACA DAN MEMFILTER DATA ---
# Membaca data CSV ke dalam DataFrame Pandas
# Jika CSV Anda menggunakan pemisah titik koma, ubah menjadi: pd.read_csv(io.BytesIO(uploaded[nama_file_awal]), sep=';')
df = pd.read_csv(io.BytesIO(uploaded[nama_file_awal]))

# Menampilkan informasi sebelum pembersihan
jumlah_baris_awal = len(df)
print(f"Jumlah baris sebelum difilter: {jumlah_baris_awal} baris")

# PROSES FILTERING (MENGHAPUS DUPLIKAT)
# Metode drop_duplicates() secara default akan menghapus baris yang semua nilai kolomnya sama persis.
df_bersih = df.drop_duplicates()

# TIPS: Jika Anda hanya ingin mengecek redundansi berdasarkan kolom tertentu (misalnya 'ID' atau 'Email'):
# df_bersih = df.drop_duplicates(subset=['ID', 'Email'], keep='first')
# Parameter keep='first' berarti menyimpan data yang pertama kali muncul. Bisa diubah ke 'last' atau False.

# Menampilkan informasi setelah pembersihan
jumlah_baris_akhir = len(df_bersih)
print(f"Jumlah baris setelah difilter: {jumlah_baris_akhir} baris")

# Menghitung selisih data
jumlah_redundan = jumlah_baris_awal - jumlah_baris_akhir
print(f"-> Berhasil menghapus {jumlah_redundan} baris data yang redundan/duplikat.\n")

# --- SEL 3: MENYIMPAN DAN MENGUNDUH HASIL ---
# Menentukan nama file baru untuk data yang sudah bersih
nama_file_baru = 'data_bersih_tanpa_redundan.csv'

# Menyimpan DataFrame kembali ke bentuk CSV (index=False agar nomor baris Pandas tidak ikut tersimpan)
df_bersih.to_csv(nama_file_baru, index=False)
print(f"Data bersih telah disimpan ke dalam file: '{nama_file_baru}'")

# Memerintahkan Google Colab untuk mengunduh file otomatis ke komputer/perangkat Anda
print("Sedang mengunduh file...")
files.download(nama_file_baru)
```
