```python
!pip install apify-client pandas openpyxl -q

from apify_client import ApifyClient
import pandas as pd
from google.colab import files

# ── ✏️  EDIT THESE ──────────────────────────────────────────
APIFY_TOKEN  = "apify_api_QgqeP0nhSM1tGssufanCstEXLqTjbj3HOYN0"
SEARCH_TERMS = ["#kecuranganpemilu", "#pemilucurang", "#dirtyvote2"]
MAX_ITEMS    = 5000
DATE_START   = "2025-10-20"
DATE_END     = "2025-10-27"
OUTPUT_FILE  = "dataset_pemilu.csv"
# ────────────────────────────────────────────────────────────

client = ApifyClient(APIFY_TOKEN)

search_with_dates = []
for term in SEARCH_TERMS:
    q = term
    if DATE_START:
        q += f" since:{DATE_START}"
    if DATE_END:
        q += f" until:{DATE_END}"
    search_with_dates.append(q)

input_data = {
    "searchTerms" : search_with_dates,
    "maxItems"    : MAX_ITEMS,
    "queryType"   : "Latest",
    "filter:nativeretweets"  : False,
    "include:nativeretweets" : False,
}

print("🚀 Mulai scraping...")
print(f"   Hashtag  : {SEARCH_TERMS}")
print(f"   Rentang  : {DATE_START} s/d {DATE_END or 'sekarang'}")
print(f"   Max item : {MAX_ITEMS:,}")

run  = client.actor("CJdippxWmn9uRfooo").call(run_input=input_data)
data = []

# ✅ FIXED: use .default_dataset_id attribute instead of run["defaultDatasetId"]
for item in client.dataset(run.default_dataset_id).iterate_items():
    text = item.get("text", "")
    data.append({
        "text"     : text,
        "date"     : item.get("createdAt", ""),
        "username" : item.get("author", {}).get("userName", ""),
        "likes"    : item.get("likeCount",   0),
        "retweets" : item.get("retweetCount", 0),
        "hashtags" : ", ".join(w for w in text.split() if w.startswith("#")),
    })

df = pd.DataFrame(data)

df["date"] = pd.to_datetime(df["date"], utc=True, format="mixed")
df["date"] = df["date"].dt.tz_convert("Asia/Jakarta")
df["date"] = df["date"].dt.strftime("%Y-%m-%d")

before = len(df)
df = df.drop_duplicates(subset=["text"])

if DATE_START:
    df = df[df["date"] >= DATE_START]
if DATE_END:
    df = df[df["date"] <= DATE_END]

df = df.sort_values("date", ascending=False).reset_index(drop=True)

print(f"\n✅ Total terkumpul  : {len(df):,} tweet")
print(f"📅 Rentang tanggal  : {df['date'].min()} s/d {df['date'].max()}")
print(f"🔁 Duplikat dihapus : {before - len(df)}")
print(f"👤 Unique user      : {df['username'].nunique():,}")
print(f"❤️  Total likes      : {df['likes'].sum():,}")

df.to_csv(OUTPUT_FILE, index=False)
print(f"\n💾 Tersimpan → {OUTPUT_FILE}")
files.download(OUTPUT_FILE)
```
